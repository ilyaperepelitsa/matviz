# The wrong way

```python
from IPython.display import display_png
import matplotlib.pyplot as plt
import pandas as pd

from matplotlib.backends.backend_agg import FigureCanvasAgg
from matplotlib.figure import Figure
```

```python
first_set = pd.read_csv("~/quant/nytaxi/fastest_routes_train_part_1.csv")
second_set = pd.read_csv("~/quant/nytaxi/fastest_routes_train_part_2.csv")
```

```python
first_set.shape
```

```python
second_set.shape
```

```python
total_set = pd.concat([first_set, second_set])
```

```python
total_set.shape
```

```python
total_set.info()
```

```python
fig = Figure(figsize = (12, 5))
canvas = FigureCanvasAgg(fig)
ax = fig.add_subplot(111)
```

```python
x = total_set["number_of_steps"]

ax.set_xlim([x.min()*0.9, x.max()*1.1])
```

```python
y = total_set["total_distance"]

ax.set_ylim([y.min()*0.9, y.max()*1.1])
```

```python
ax.scatter(x, y)
```

```python
display_png(fig)
```

```python
x = total_set["number_of_steps"]

ax.set_xlim([x.min(), x.max()])

y = total_set["total_distance"]

ax.set_ylim([y.min(), y.max()])

ax.scatter(x, y)

display_png(fig)
```

# The right way

```python
import matplotlib.pyplot as plt
```

```python
x = total_set["number_of_steps"]
y = total_set["total_distance"]
```

```python
plt.figure(figsize = (10, 5))
plt.scatter(x, y)
plt.xlim(x.min(), x.max())
plt.ylim(y.min(), y.max())
plt.title("Trips and things")
plt.show()
```

# PLT Tutorial

```python
import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats
import numpy as np
set_matplotlib_formats("retina")
```

```python
for i in range(3):
    plt.plot(np.random.randn(10))

plt.title("Normally distributed Randon Samples")
plt.show()
```

```python
l1, l2, l3 = np.random.randn(3, 20)
plt.plot(l1)
plt.show()
```

```python
xs = range(10, 10 + len(l1))
plt.plot(xs, l1)
plt.show()
```

```python
plt.plot(l1, "r--s")
plt.show()
```

```python
plt.plot(l1, "", l2, "", l3, "")
plt.show()
plt.rcParams["axes.prop_cycle"]
```

```python
# even number of lists, each pair - x and y

plt.plot(range(len(l1)), l1,
        range(len(l2)), l2,
        range(len(l3)), l3)
plt.show()
```
